```python
u="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
p=req.get(u)
c=BeautifulSoup(p.text,'html')
print(c)
```

```python
h=a.find_all('th')
h
```

```python
heading=[i.text for i in h]
heading
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=heading)
print(mydf)
```

```python
data_act=[i.text for i in data]
print(data_act)
```

```python
df1=pd.DataFrame(columns=['name','rank','nw','comp'])
df1
```

```python
headings=[i.text for i in p]
print(headings)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=headings)
(df)
```

```python
data_act1=[i.text for i in row_data]
print(data_act1)
```

```python
print(len(data_act1))
```
